```python
# Imports
import pandas as pd
import openpyxl
```

```python
# Load the Excel file
file_path = '拆表已签-中移铁通（福州）（2024-2025）202409——全日制——建设.xlsx'
sheet_name = 0  # Select the first sheet in the Excel file

df = pd.read_excel(file_path, sheet_name=sheet_name, engine='openpyxl')
```

```python
df['Unnamed: 4']
```

```python
# Set column name to look for
salary_column = '应付工资'
name_column = '姓名'
bank_account_column = '银行账号'
target_columns = [name_column, bank_account_column, salary_column]
```

```python
# compiling new dataframe

columns_found = {}
remaining_columns = target_columns.copy()
max_row_index = min(10, len(df))
largest_data_start_row = 0

# First pass: Find the columns and the largest data_start_row
for row_index in range(max_row_index):
    for col_index in range(df.shape[1]):
        cell_value = df.iloc[row_index, col_index]
        if cell_value in remaining_columns:
            data_start_row = row_index + 1
            largest_data_start_row = max(largest_data_start_row, data_start_row)
            columns_found[cell_value] = col_index
            remaining_columns.remove(cell_value)
            if not remaining_columns:
                break
    if not remaining_columns:
        break

# Second pass: Extract data starting from the largest_data_start_row
for col_name, col_index in columns_found.items():
    data_series = df.iloc[largest_data_start_row:, col_index].reset_index(drop=True)
    columns_found[col_name] = data_series

new_df = pd.DataFrame(columns_found)
```

```python
# Post processing

new_df = new_df.dropna()
new_df = new_df[~new_df['姓名'].apply(lambda x: str(x).isdigit())]
new_df = new_df.reset_index(drop=True)
```

```python
# Data validation

new_df = new_df[target_columns]
new_df
```

```python
# Export the new DataFrame to Excel
output_file_path = 'processed_salary_data.xlsx'
new_df.to_excel(output_file_path, index=False)

print(f"Data exported successfully to {output_file_path}")
```
